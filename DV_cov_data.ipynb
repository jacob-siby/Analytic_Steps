```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=data['Date']
y=data['Confirmed']
```

```python
fig, xy = plt.subplots()  # Create a figure containing a single axes.
xy.plot(x, y)
#plt.bar(x, y, color = 'r', width = 1, label = "DEATH")
#plt.subplot(2, 100, 1)
plt.xlabel('Date')
plt.ylabel('No: of death')
plt.title('Confirmed Cases')
plt.legend()
plt.show()
fig.savefig('test.png')
```

```python
x=data['State/UnionTerritory']
y=data['Deaths']
plt.bar(x, y, color = 'r', width = 1, label = "DEATH")
#plt.subplot(2, 100, 1)
plt.xlabel('State/UnionTerritory')
plt.ylabel('No: of death')
plt.title('Deaths by Covid19')
plt.legend()

plt.show()
```

#STATE - DEATHS

```python

x=data['State/UnionTerritory']
y=data['Deaths']

  
# plotting a line plot after changing it's width and height
f = plt.figure()
f.set_figwidth(70)
f.set_figheight(70)
#plt.legend()
plt.barh(x, y)


plt.legend()
plt.title('State/UnionTerritory VS Deaths', fontsize=65, fontweight='bold')
plt.xlabel('State/UnionTerritory', fontsize=40)
plt.ylabel('Deaths', fontsize=40) 

plt.yticks(fontsize=30)
plt.xticks(fontsize=30)

#plt.barh(x, y)

plt.show()

#plt.show()
f.savefig('state_death_brah.png')
```

# State/UnionTerritory - Confirmed

```python

x=data['State/UnionTerritory']
y=data['Confirmed']
  
  
# plotting a line plot after changing it's width and height
p = plt.figure()
p.set_figwidth(70)
p.set_figheight(70)

plt.barh(x, y)
plt.legend()

plt.title('State/UnionTerritory VS Confirmed Cases', fontsize=65, fontweight='bold')
plt.xlabel('State/UnionTerritory', fontsize=40)
plt.ylabel('Confirmed Cases', fontsize=40)
  

plt.yticks(fontsize=30)
plt.xticks(fontsize=30)

plt.show()
p.savefig('test2.png')
```
